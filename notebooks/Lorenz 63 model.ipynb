```python
import numpy as np
import dapy.inference as da
from dapy.models.lorenz63 import Lorenz63Model
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
%matplotlib inline
sns.set_style('white')
```

# Model

Three-dimensional dynamical system with flow map defined by ODEs

\begin{align}
  \frac{\mathrm{d} x_0}{\mathrm{d} t} &= \sigma (x_1 - x_0) \\
  \frac{\mathrm{d} x_1}{\mathrm{d} t} &= x_0 (\rho - x_2) - x_1 \\
  \frac{\mathrm{d} x_2}{\mathrm{d} t} &= x_0 x_1 - \beta x_2. \\
\end{align}

For parameter values $\rho = 28, \sigma=10, \beta=\frac{8}{3}$ exhibits chaotic dynamics.

Originally proposed in

> Lorenz, Edward Norton (1963).
> Deterministic nonperiodic flow.
> Journal of the Atmospheric Sciences. 20 (2): 130â€“141.

```python
n_steps = 400
seed = 20171027
rng = np.random.RandomState(seed)
observation_func = lambda z, t: z.T[0:1].T
state_noise_std=1.
obser_noise_std = 5.
model = Lorenz63Model(
    rng=rng, state_noise_std=state_noise_std, obser_noise_std=obser_noise_std,
    dt=0.005, observation_func=observation_func)
```

## Generate data from model

```python
z_reference, x_reference = model.generate(n_steps)
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(z_reference)
ax.set_xlabel('Time index $t$')
_ = ax.set_xlim(0, n_steps - 1)
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_reference)
ax.set_xlabel('Time index $t$')
_ = ax.set_xlim(0, n_steps - 1)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = Axes3D(fig, [0, 0, 1, 1])
ax.plot(z_reference[:, 0], z_reference[:, 1], z_reference[:, 2], '-', lw=1)
```

## Infer state from observations

### Ensemble Kalman filter (perturbed observations)

```python
enkf = da.EnsembleKalmanFilter(
    model.init_state_sampler, model.next_state_sampler, 
    model.observation_sampler, rng
)
```

```python
%%time
results_enkf = enkf.filter(x_reference, 100)
```

```python
for i in range(3):
    fig = plt.figure(figsize=(12, 4))
    ax = fig.add_subplot(1, 1, 1)
    _ = ax.plot(results_enkf['z_particles_seq'][:, :, i], 'r-', lw=0.25, alpha=0.25, label='EnkF')
    _ = ax.plot(z_reference[:, i], 'k--', label='True')
```

### Bootstrap particle filter

```python
bspf = da.BootstrapParticleFilter(
    model.init_state_sampler, model.next_state_sampler, 
    model.log_prob_dens_obs_gvn_state, rng
)
```

```python
%%time
results_bspf = bspf.filter(x_reference, 100)
```

```python
for i in range(3):
    fig = plt.figure(figsize=(12, 4))
    ax = fig.add_subplot(1, 1, 1)
    _ = ax.plot(results_bspf['z_particles_seq'][:, :, i], 'r-', lw=0.25, alpha=0.25, label='EnkF')
    _ = ax.plot(z_reference[:, i], 'k--', label='True')
```

### Ensemble transform particle filter

```python
etpf = da.EnsembleTransformParticleFilter(
    model.init_state_sampler, model.next_state_sampler, 
    model.log_prob_dens_obs_gvn_state, rng
)
```

```python
%%time
results_etpf = etpf.filter(x_reference, 100)
```

```python
for i in range(3):
    fig = plt.figure(figsize=(12, 4))
    ax = fig.add_subplot(1, 1, 1)
    _ = ax.plot(results_etpf['z_particles_seq'][:, :, i], 'r-', lw=0.25, alpha=0.25, label='EnkF')
    _ = ax.plot(z_reference[:, i], 'k--', label='True')
```

### Visualise estimated means of filtering distribution

```python
for i in range(3):
    fig = plt.figure(figsize=(12, 4))
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(z_reference[:, i], 'k', label='True')
    ax.plot(results_enkf['z_mean_seq'][:, i], ':', label='EnKF')
    ax.plot(results_bspf['z_mean_seq'][:, i], '--', label='BSPF')
    ax.plot(results_etpf['z_mean_seq'][:, i], '-.', label='ETPF')
    ax.legend(ncol=4)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = Axes3D(fig, [0, 0, 1, 1])
ax.plot(
    results_enkf['z_mean_seq'][:, 0], 
    results_enkf['z_mean_seq'][:, 1], 
    results_enkf['z_mean_seq'][:, 2], 
    ':', lw=1, label='EnKF')
ax.plot(
    z_reference[:, 0], 
    z_reference[:, 1], 
    z_reference[:, 2], 
    'k-', lw=1, label='True')
ax.set_title('EnKF')
fig = plt.figure(figsize=(8, 8))
ax = Axes3D(fig, [0, 0, 1, 1])
ax.plot(
    results_bspf['z_mean_seq'][:, 0], 
    results_bspf['z_mean_seq'][:, 1], 
    results_bspf['z_mean_seq'][:, 2], 
    '--', lw=1, label='BSPF')
ax.plot(
    z_reference[:, 0], 
    z_reference[:, 1], 
    z_reference[:, 2], 
    'k-', lw=1, label='True')
ax.set_title('BSPF')
fig = plt.figure(figsize=(8, 8))
ax = Axes3D(fig, [0, 0, 1, 1])
ax.plot(
    results_etpf['z_mean_seq'][:, 0], 
    results_etpf['z_mean_seq'][:, 1], 
    results_etpf['z_mean_seq'][:, 2], 
    '-.', lw=1, label='ETPF')
ax.plot(
    z_reference[:, 0], 
    z_reference[:, 1], 
    z_reference[:, 2], 
    'k-', lw=1, label='True')
ax.set_title('ETPF')
```
